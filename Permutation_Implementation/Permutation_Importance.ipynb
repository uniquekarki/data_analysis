```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
import eli5
from eli5.sklearn import PermutationImportance
from sklearn.feature_selection import SelectFromModel
```

```python
train_data=pd.read_csv('train.csv')
test_data=pd.read_csv('test.csv')
print('The dataframe is read')
```

```python
train_data.head()
```

```python
target = 'SalePrice'
```

```python
X_train = train_data.select_dtypes(include = ['number']).copy()
X_train = X_train.drop([target],axis=1)
y_train = train_data[target]
X_test = test_data.select_dtypes(include = ['number']).copy()
```

```python
X_train = X_train.fillna(X_train.mean())
X_test = X_test.fillna(X_test.mean())
```

```python
type(X_test)
```

```python
train_X, val_X, train_y, val_y = train_test_split(X_train, y_train, random_state=1)
```

```python
regressor = RandomForestRegressor(n_estimators = 100, max_depth = 10)
regressor.fit(train_X,train_y)
```

```python
perm = PermutationImportance(regressor, random_state = 1)
perm.fit(val_X, val_y)
eli5.show_weights(perm, feature_names = val_X.columns.tolist())
```

```python
features = ['OverallQual','GrLivArea','2ndFlrSF','BsmtFinSF1','TotalBsmtSF','GarageCars','GarageArea','LotArea']
```

```python
train_X = train_X[features]
X_test = X_test[features]
```

```python
regressor.fit(train_X, train_y)
```

```python
predictions = regressor.predict(X_test)
```

```python
output = pd.DataFrame({"Id":test_data.index, target:predictions})
output.to_csv('Permutation_result.csv', index=False)
```
